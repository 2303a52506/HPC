```python
%%bash
# Save the MPI C code to a file
cat > hello.c << EOF
#include <mpi.h>
#include <stdio.h>

int main(int argc, char *argv[]) {
    MPI_Init(&argc, &argv);

    int rank, size;
    MPI_Comm_rank(MPI_COMM_WORLD, &rank);
    MPI_Comm_size(MPI_COMM_WORLD, &size);
    printf("Hello from process %d out of %d processes\n", rank, size);
    MPI_Finalize();
    return 0;
}
EOF

# Compile the MPI C code
mpicc hello.c -o hello
# Run the MPI program with 4 processes, using --oversubscribe to ignore slot limitations
mpirun --allow-run-as-root --oversubscribe -np 4 ./hello
```

```python
%%bash
# Save the MPI C code to a file
cat > send.c << EOF
#include <mpi.h>
#include <stdio.h>
#include <string.h>

int main(int argc, char *argv[]) {
    MPI_Init(&argc, &argv);

    int rank;
    char message[100];

    MPI_Comm_rank(MPI_COMM_WORLD, &rank);

    if(rank == 0) {
        strcpy(message, "Hello from Process 0");
        MPI_Send(message, strlen(message)+1, MPI_CHAR, 1, 0, MPI_COMM_WORLD);
    } else if(rank == 1) {
        MPI_Recv(message, 100, MPI_CHAR, 0, 0, MPI_COMM_WORLD, MPI_STATUS_IGNORE);
        printf("Received: %s\n", message);
    }

    MPI_Finalize();
    return 0;
}
EOF

# Compile the MPI C code
mpicc send.c -o send

# Execute the MPI program
mpirun --allow-run-as-root --oversubscribe -np 2 ./send
Program 3: Broadcast
%%bash
# Save the MPI C code to a file
cat > send.c << EOF
#include <mpi.h>
#include <stdio.h>
#include <string.h>

int main(int argc, char *argv[]) {
    MPI_Init(&argc, &argv);

    int rank;
    char message[100];

    MPI_Comm_rank(MPI_COMM_WORLD, &rank);

    if(rank == 0) {
        strcpy(message, "Hello from Process 0");
        MPI_Send(message, strlen(message)+1, MPI_CHAR, 1, 0, MPI_COMM_WORLD);
    } else if(rank == 1) {
        MPI_Recv(message, 100, MPI_CHAR, 0, 0, MPI_COMM_WORLD, MPI_STATUS_IGNORE);
        printf("Received: %s\n", message);
    }

    MPI_Finalize();
    return 0;
}
EOF

# Compile the MPI C code
mpicc send.c -o send

# Execute the MPI program
mpirun --allow-run-as-root --oversubscribe -np 2 ./send
Program 4: Reduce (Sum)
%%bash
# Save the MPI C code to a file
cat > reduce.c << EOF
#include <mpi.h>
#include <stdio.h>

int main(int argc, char *argv[]) {
    MPI_Init(&argc, &argv);

    int rank, sum;
    int value = 1;

    MPI_Comm_rank(MPI_COMM_WORLD, &rank);

    MPI_Reduce(&value, &sum, 1, MPI_INT, MPI_SUM, 0, MPI_COMM_WORLD);

    if(rank == 0) {
        printf("Sum = %d\n", sum);
    }

    MPI_Finalize();
    return 0;
}
EOF

# Compile the MPI C code
mpicc reduce.c -o reduce

# Execute the MPI program
mpirun --allow-run-as-root --oversubscribe -np 4 ./reduce
```

```python
%%bash
# Save the MPI C code to a file
cat > send_recv_pairs.c << EOF
#include <mpi.h>
#include <stdio.h>

int main(int argc, char *argv[]) {
    int rank, size;
    int data;

    MPI_Init(&argc, &argv);
    MPI_Comm_rank(MPI_COMM_WORLD, &rank);
    MPI_Comm_size(MPI_COMM_WORLD, &size);

    // Ensure even number of processes
    if (size % 2 != 0) {
        if (rank == 0) {
            printf("Please run with an even number of processes.\n");
        }
        MPI_Finalize();
        return 0;
    }

    if (rank % 2 == 0) {
        // Even-ranked process sends data
        data = rank * 10;  // example data
        printf("Process %d sending data %d to process %d\n", rank, data, rank + 1);
        MPI_Send(&data, 1, MPI_INT, rank + 1, 0, MPI_COMM_WORLD);
    }
    else {
        // Odd-ranked process receives data
        MPI_Recv(&data, 1, MPI_INT, rank - 1, 0, MPI_COMM_WORLD, MPI_STATUS_IGNORE);
        printf("Process %d received data %d from process %d\n", rank, data, rank - 1);
    }

    MPI_Finalize();
    return 0;
}
EOF

# Compile the MPI C code
mpicc send_recv_pairs.c -o send_recv_pairs

# Execute the MPI program with an even number of processes (e.g., 4)
mpirun --allow-run-as-root --oversubscribe -np 4 ./send_recv_pairs
```
