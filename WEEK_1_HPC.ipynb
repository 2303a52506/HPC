```python

import os
import socket
import time
from datetime import datetime

print("=== Serial Python Batch Job Started ===")
print(f"Start Time      : {datetime.now()}")
print(f"Hostname        : {socket.gethostname()}")
print(f"Current User    : {os.getenv('USER')}")

# Scheduler-specific environment variables
print("\n--- Scheduler Metadata ---")
print(f"SLURM Job ID    : {os.getenv('SLURM_JOB_ID')}")
print(f"PBS Job ID      : {os.getenv('PBS_JOBID')}")
print(f"Job Name        : {os.getenv('SLURM_JOB_NAME') or os.getenv('PBS_JOBNAME')}")
print(f"Working Dir     : {os.getcwd()}")

print("\n--- Job Execution ---")
for i in range(1, 6):
    print(f"Processing step {i}/5 ...")
    time.sleep(1)

print("\n=== Job Completed Successfully ===")
print(f"End Time        : {datetime.now()}")
```
